```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('output/wikimedia_events_combined.csv')
df
```

```python
sns.scatterplot(data=df, x='length_diff', y='length_diff_percent')
```

```python
max_length_diff_users = df.groupby('user')['length_diff'].max().reset_index().sort_values(by='length_diff', ascending=False).head(5)
sns.barplot(data=max_length_diff_users, x='user', y='length_diff')
plt.title('Top 5 Users by Maximum Length Difference')
plt.ylabel('Length Difference')
plt.xlabel('User')
```

```python
max_length_diff_domain = df.groupby('domain')['length_diff'].max().reset_index().sort_values(by='length_diff', ascending=False).head(5)
sns.barplot(data=max_length_diff_domain, x='domain', y='length_diff')
plt.title('Top 5 Users by Maximum Length Difference Percentage')
plt.ylabel('Length Difference')
plt.xlabel('Domain')
```

```python
top_domains_bot = df.groupby('domain')['bot'].max().reset_index().sort_values(by='bot', ascending=False).head(5)
sns.barplot(data=top_domains_bot, x='domain', y='bot')
plt.title('Bots by top 5 domains')
plt.ylabel('Bot')
plt.xlabel('Domain')
```
